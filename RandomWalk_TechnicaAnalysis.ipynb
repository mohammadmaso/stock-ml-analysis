```
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go


```

```
for x in range(10):
  y = np.cumsum(np.random.normal(size=1000))
  a = y[:500]
  none = np.full((500), None)


  series1 = np.concatenate((a , none),axis=None)
  series2 = y
  x1 = list(range(0,500))
  x2 = list(range(0,1000))

  df = pd.DataFrame()
  df['id']=x2
  df["y1"] = series1
  df['y2'] = series2
  #print(df.head())
  colors = px.colors.qualitative.Plotly
  fig = go.Figure()
  fig.add_traces(go.Scatter(x=df['id'], y = df['y2'], mode = 'lines', line=dict(color=colors[1])))
  fig.add_traces(go.Scatter(x=df['id'], y = df['y1'], mode = 'lines', line=dict(color=colors[0])))

  fig.show()#fig = px.line( df, y=['y1', 'y2'])
```
